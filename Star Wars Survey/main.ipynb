### Overview

- Reading the data set into a dataframe, with proper encoding and non-null values.

```python
import pandas as pd
star_wars = pd.read_csv('StarWars.csv', encoding='ISO-8859-1')
star_wars = star_wars[star_wars['RespondentID'].notnull()]

pd.set_option('display.max_columns', 200)
star_wars.head()
```

### Cleaning and Mapping Yes/No Columns

- Converting the first two columns into Boolean type containing only the values - True, False, and NaN.

```python
map_dict = {'Yes': True, 'No': False}

cols = star_wars.columns[1:3]

for i in cols:
    star_wars[i] = star_wars[i].map(map_dict)

star_wars
```

### Cleaning and Mapping Movie Seen Columns

- Converting column[3:9] so that it only contains the values True and False, instead of movie names.<br>
- Renaming each of the columns above, so the names are more intuitive. For e.g. seen_1 to indicate whether the respondent saw Star Wars: Episode I The Phantom Menace, seen_2 for Star Wars: Episode II Attack of the Clones, and so on.

```python
import numpy as np

col_to_rename = star_wars.columns[3:9]


naming_dict = {'Star Wars: Episode I  The Phantom Menace': True,
               'Star Wars: Episode II  Attack of the Clones': True,
               'Star Wars: Episode III  Revenge of the Sith': True,
               'Star Wars: Episode IV  A New Hope': True,
               'Star Wars: Episode V The Empire Strikes Back': True,
               'Star Wars: Episode VI Return of the Jedi': True,
               np.NaN: False}

for i in col_to_rename:
    star_wars[i] = star_wars[i].map(naming_dict)
    
star_wars = star_wars.rename(columns={col_to_rename[0] : 'seen_1',
                                     col_to_rename[1] : 'seen_2',
                                     col_to_rename[2] : 'seen_3',
                                     col_to_rename[3] : 'seen_4',
                                     col_to_rename[4] : 'seen_5',
                                     col_to_rename[5] : 'seen_6'})
star_wars.head()
```

### Cleaning the Ranking Columns

- Converting columns[9:15] to a float type.
- Renaming each column to a more descriptive name. For e.g. ranking_1, ranking_2, and so on.

```python
star_wars.iloc[:,9:15] = star_wars.iloc[:,9:15].astype(float)

col_to_rename = star_wars.columns[9:15]
# col_to_rename
star_wars = star_wars.rename(columns={col_to_rename[0] : 'ranking_1',
                                     col_to_rename[1] : 'ranking_2',
                                     col_to_rename[2] : 'ranking_3',
                                     col_to_rename[3] : 'ranking_4',
                                     col_to_rename[4] : 'ranking_5',
                                     col_to_rename[5] : 'ranking_6'})

star_wars.head()
```

### Finding The Highest-Ranked Movie

```python
import matplotlib.pyplot as plt
%matplotlib inline

print(star_wars[star_wars.columns[9:15]].mean())

index = np.arange(6) 
movie_labels = ['Episode I',
               'Episode II',
               'Episode III',
               'Episode IV',
               'Episode V',
               'Episode VI']

plt.bar(index, star_wars[star_wars.columns[9:15]].mean())
plt.xticks(index, movie_labels, rotation=80)
plt.show()
```

**Since, a lower ranking is better. We can clearly observe that the original movies (which were released first) are the most liked ones. And the Prequels which were made afterwards are not.**

### Explore the Data by Gender

We will split dataframe into two groups based on `Gender` by creating two subsets of that column. Like this:

`males = star_wars[star_wars["Gender"] == "Male"]`<br>
`females = star_wars[star_wars["Gender"] == "Female"]`

```python
males = star_wars[star_wars['Gender']=='Male']
females = star_wars[star_wars['Gender']=='Female']

fig = plt.figure(figsize=(14, 8))

for x in range(1, 5):
    ax = fig.add_subplot(2, 2, x)
    ax.spines["right"].set_visible(False)    
    ax.spines["top"].set_visible(False)    
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    
    if x < 3:
        if x == 1:
            ax.bar(index, males[males.columns[9:15]].mean())
            ax.set_title('Male Rankings')
        else:
            ax.bar(index, females[females.columns[9:15]].mean())
            ax.set_title('Female Rankings')
        ax.set_yticks(np.arange(5))
        ax.set_yticklabels(np.arange(5))
        
    else:
        if x == 3:
            ax.bar(index, males[males.columns[3:9]].mean())
            ax.set_title('Male Views')
        else:
            ax.bar(index, females[females.columns[3:9]].mean())
            ax.set_title('Female Views')
        ax.tick_params(labelbottom=True)
        ax.set_ylim(0, 0.8)
        ax.set_yticks(np.arange(0, 1, 0.2))
        ax.set_yticklabels(np.arange(0, 1, 0.2).round(decimals=2))
        ax.set_xticks(index)
        ax.set_xticklabels(movie_labels, rotation=80)
            
plt.show()
```

**Here, we have an interesting observation that more males watched the newer movies, but they liked it lesser then their female counterparts.**
